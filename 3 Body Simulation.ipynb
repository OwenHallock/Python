In this Lab we will examine the motion of a 3rd body small mass object in the Earth-Moon system.

We will examine two scenarios:

- a rocket launched from Earth towards the Moon
- an asteroid entering the Earth-Moon system and determining if it will hit Earth


For the first scenario of the rocket, you are tasked to dermine the correct velocity and angle so that the trajectory of the rocket brings it close to the moon without hitting it, and then returns to the Earth. 

For the Asteroid, given the intial velocity and direction, you are tasked to determine what initial position of the moon would cause the asteroid to impact earth. And, if it does impact how much kineic energy it will have.

The equations that govern the motion are from Newton's law and the law of gravity:

$\;\;\;\;\; \vec{F}=m\vec{a}=m\ddot{\vec{r}} \;\;\;\;\; \vec{F}_{Grav} = -\frac{GMm}{r^2}\hat{r}=-\frac{GMm}{r^3}\vec{r}$

where $M,m$ are the masses of 2 objects, $r$ is the distance seperating them, $G$ is Newton's gravitational constant, and $\vec{r}$ is the vector from one object to the other.

For our problem, we will assume all three objects are in a plane, this becomes a 2-d problem, for which we will use cartesian coordinates (x,y) with the center of the Earth $M$ as the origin. We will label the moon object with $m$, and the rocket(or asteroid) object with $R$.

Earth to moon vector $=\vec{r_m}$ <br>
Earth to object vector $=\vec{r_R}$<br>
Moon to object vector $=\vec{r_{mR}}=\vec{r_R}-\vec{r_m}$

We will assume the moon and 3rd object's masses are small enough such that 
- the Earth stays at the origin
- the Moon's motion is unpertubed by the 3rd object

Under these conventions and assumption we have the 2 second order 2D vector ODEs

$m_m\ddot{\vec{r_m}} = -\frac{GMm_m}{r_m^3}\vec{r_m}$

$m_R\ddot{\vec{r_R}} = -\frac{GMm_R}{r_R^3}\vec{r_R} -\frac{Gm_mm_R}{r_{mR}^3}\vec{r_{mR}} $


Then, using $\vec{r}=x\hat{i}+y\hat{j}$ and canceling of $m_m,m_R$ the vector equations above can be written as:

$\ddot{x_m}\hat{i}+\ddot{y_m}\hat{j}=-\frac{GM}{(x_m^2+y_m^2)^{3/2}}(x_m\hat{i}+y_m\hat{j})$ 

$\ddot{x_R}\hat{i}+\ddot{y_R}\hat{j}=-\frac{GM}{(x_R^2+y_R^2)^{3/2}}(x_R\hat{i}+y_R\hat{j})
- \frac{Gm_m}{((x_m-x_R)^2+(y_m-y_R)^2)^{3/2}}((x_R-x_m)\hat{i}+(y_R-y_m)\hat{j})$ 

or writing the equations for each vector component we get 4 coupled second order ODEs,

$\ddot{x_m}=-\frac{GM}{(x_m^2+y_m^2)^{3/2}}x_m$ <br>

$\ddot{y_m}=-\frac{GM}{(x_m^2+y_m^2)^{3/2}}y_m$ <br>

$\ddot{x_R}=-\frac{GM}{(x_R^2+y_R^2)^{3/2}}x_R - \frac{Gm_m}{((x_R-x_m)^2+(y_R-y_m)^2)^{3/2}}(x_R-x_m)$<br>

$\ddot{y_R}=-\frac{GM}{(x_R^2+y_R^2)^{3/2}}y_R - \frac{Gm_m}{((x_R-x_m)^2+(y_R-y_m)^2)^{3/2}}(y_R-y_m) $ <br>



As we did in the class examples, we can break the 2nd oder ODEs into coupled 1st order ODEs. In this case we have 4 coupled 2nd order ODEs to be written as 8 couple 1st order ODEs. NOTE: I will start subscripts at 0 to correspond to the correct array index in python.

We have 8 variables

$y_0=x_m$<br>
$y_1=\dot{x}_m = \dot{y_0}$<br>
$y_2=y_m$ <br>
$y_3=\dot{y}_m = \dot{y_2}$<br>
$y_4=x_R$<br>
$y_5=\dot{x}_R = \dot{y_4}$<br>
$y_6=y_R$ <br>
$y_7=\dot{y}_R = \dot{y_6}$<br>

Then the 8 coupled 1st order ODEs are:

$\dot{y_0} = \dot{x_m}=y_1$ <br>
$\dot{y_1} = \ddot{x_m} = -\frac{GM}{(x_m^2+y_m^2)^{3/2}}x_m = -\frac{GM}{(y_0^2+y_2^2)^{3/2}}y_0$<br>
$\dot{y_2} = \dot{y_m}=y_3$<br>
$\dot{y_3} = \ddot{y_m} = -\frac{GM}{(x_m^2+y_m^2)^{3/2}}y_m = -\frac{GM}{(y_0^2+y_2^2)^{3/2}}y_2$<br> 
$\dot{y_4} = \dot{x_R}=y_5$ <br>
$\dot{y_5} = \ddot{x_R} = -\frac{GM}{(x_R^2+y_R^2)^{3/2}}x_R - \frac{Gm_m}{((x_R-x_m)^2+(y_R-y_m)^2)^{3/2}}(x_R-x_m)= -\frac{GM}{(y_4^2+y_6^2)^{3/2}}y_4 -\frac{Gm_m}{((y_4-y_0)^2+(y_6-y_2)^2)^{3/2}}(y_4-y_0)$
<br>
$\dot{y_6} = \dot{y_R}=y_7$<br>
$\dot{y_7} = \ddot{y_R} = -\frac{GM}{(x_R^2+y_R^2)^{3/2}}y_R- \frac{Gm_m}{((x_R-x_m)^2+(y_R-y_m)^2)^{3/2}}(y_R-y_m)= -\frac{GM}{(y_4^2+y_6^2)^{3/2}}y_6-\frac{Gm_m}{((y_4-y_0)^2+(y_6-y_2)^2)^{3/2}}(y_6-y_2)$<br> 



Putting this in array form for use with odeint in python:

$\;\;\;\; \mathbf{y} = \left[ \begin{array}{c} y_0 \\ y_1 \\y_2 \\y_3\\y_4\\y_5\\y_6\\y_7 \end{array} \right]
\;\;\;\;\;\;\;\; \frac{d\mathbf{y}}{dt} = \left[ \begin{array}{c} y_1 \\ -\frac{GM}{(y_0^2+y_2^2)^{3/2}}y_0\\ y_3 \\ -\frac{GM}{(y_0^2+y_2^2)^{3/2}}y_2 \\ y_5 \\ -\frac{GM}{(y_4^2+y_6^2)^{3/2}}y_4 -\frac{Gm_m}{((y_4-y_0)^2+(y_6-y_2)^2)^{3/2}}(y_4-y_0) \\y_7\\ -\frac{GM}{(y_4^2+y_6^2)^{3/2}}y_6-\frac{Gm_m}{((y_4-y_0)^2+(y_6-y_2)^2)^{3/2}}(y_6-y_2)
\end{array} \right]$

where the second equation for the 1st derivative is what you write as your user function "def F(y,t):" that odeint calls. (see examples from class on ELMS).
